```python
import beahvioural_functions as bf
import pingouin as pin
import pandas as pd
import ptitprince as pt
from IPython.display import display
from statsmodels.stats.multitest import multipletests as multitest
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
```

```python
tasks = {'eft_simple': bf.get_mean_rt_df('eft', 'SimpleFigures'), 
         'eft_complex': bf.get_mean_rt_df('eft', 'ComplexFigures'), 
         'fear_faces': bf.get_mean_rt_df('fear', 'fear'),
         'partial_fear': bf.get_mean_rt_df('fear', 'Partially_fear'),
         'happy_faces': bf.get_mean_rt_df('happy', 'happy'),
         'partial_happy': bf.get_mean_rt_df('happy', 'Partially_happy'),
}

is_correct = {'eft_simple': bf.iscorrect('eft', 'SimpleFigures'), 
         'eft_complex': bf.iscorrect('eft', 'ComplexFigures'), 
         'fear_faces': bf.iscorrect('fear', 'fear'),
         'partial_fear': bf.iscorrect('fear', 'Partially_fear'),
         'happy_faces': bf.iscorrect('happy', 'happy'),
         'partial_happy': bf.iscorrect('happy', 'Partially_happy'),
}
```

```python
for task in tasks.keys():
    print(task)
    pt.RainCloud(
                x="time", 
                y='rt', 
                hue="group", 
                data=tasks[task], 
                palette='Accent_r', 
                alpha=.65, 
                dodge=True, 
                scale="area",
                width_viol=.5,
                jitter=0,
                pointplot = True,
                bw=.3,
                
                 )
    plt.show()
    pt.RainCloud(
                x="time", 
                y='iscorrect', 
                hue="group", 
                data=is_correct[task], 
                palette='Accent_r', 
                alpha=.65, 
                dodge=True, 
                scale="area",
                width_viol=.5,
                jitter=0,
                pointplot = True,
                bw=.3,
                
                 )
    plt.show()
```

```python
for df in tasks.keys():
    print(df)
    display(pd.DataFrame({
        'an_t1_mean': tasks[df][(tasks[df]['time'] == 't1') & (tasks[df]['group'] == 'AN')].mean(),
        'an_t1_std': tasks[df][(tasks[df]['time'] == 't1') & (tasks[df]['group'] == 'AN')].std(),
        'an_t2_mean': tasks[df][(tasks[df]['time'] == 't2') & (tasks[df]['group'] == 'AN')].mean(),
        'an_t2_std': tasks[df][(tasks[df]['time'] == 't2') & (tasks[df]['group'] == 'AN')].std(),
        'hc_t1_mean': tasks[df][(tasks[df]['time'] == 't1') & (tasks[df]['group'] == 'HC')].mean(),
        'hc_t1_std': tasks[df][(tasks[df]['time'] == 't1') & (tasks[df]['group'] == 'HC')].std(),
        'hc_t2_mean': tasks[df][(tasks[df]['time'] == 't2') & (tasks[df]['group'] == 'HC')].mean(),
        'hc_t2_std': tasks[df][(tasks[df]['time'] == 't2') & (tasks[df]['group'] == 'HC')].std(),
        }
    ).drop('sub', axis=0)
    )
```

```python
for df in is_correct.keys():
    print(df)
    display(pd.DataFrame({
        'an_t1_mean': is_correct[df][(is_correct[df]['time'] == 't1') & (is_correct[df]['group'] == 'AN')].mean(),
        'an_t1_std': is_correct[df][(is_correct[df]['time'] == 't1') & (is_correct[df]['group'] == 'AN')].std(),
        'an_t2_mean': is_correct[df][(is_correct[df]['time'] == 't2') & (is_correct[df]['group'] == 'AN')].mean(),
        'an_t2_std': is_correct[df][(is_correct[df]['time'] == 't2') & (is_correct[df]['group'] == 'AN')].std(),
        'hc_t1_mean': is_correct[df][(is_correct[df]['time'] == 't1') & (is_correct[df]['group'] == 'HC')].mean(),
        'hc_t1_std': is_correct[df][(is_correct[df]['time'] == 't1') & (is_correct[df]['group'] == 'HC')].std(),
        'hc_t2_mean': is_correct[df][(is_correct[df]['time'] == 't2') & (is_correct[df]['group'] == 'HC')].mean(),
        'hc_t2_std': is_correct[df][(is_correct[df]['time'] == 't2') & (is_correct[df]['group'] == 'HC')].std(),
        }
    ).drop('sub', axis=0)
    )
```

```python
mwu_models = {}
for measure in is_correct.keys():
    df_an_t1_correct = is_correct[measure][(is_correct[measure]['group'] == 'AN') & 
                                        (is_correct[measure]['time'] == 't1')].reset_index(drop=True)
    df_an_t2_correct = is_correct[measure][(is_correct[measure]['group'] == 'AN') & 
                                        (is_correct[measure]['time'] == 't2')].reset_index(drop=True)
    
    df_hc_t1_correct = is_correct[measure][(is_correct[measure]['group'] == 'HC') & 
                                        (is_correct[measure]['time'] == 't1')].reset_index(drop=True)
    df_hc_t2_correct = is_correct[measure][(is_correct[measure]['group'] == 'HC') & 
                                        (is_correct[measure]['time'] == 't2')].reset_index(drop=True)
    df_an_t1_rt = tasks[measure][(tasks[measure]['group'] == 'AN') & 
                                        (tasks[measure]['time'] == 't1')].reset_index(drop=True)
    df_an_t2_rt = tasks[measure][(tasks[measure]['group'] == 'AN') & 
                                        (tasks[measure]['time'] == 't2')].reset_index(drop=True)
    
    df_hc_t1_rt = tasks[measure][(tasks[measure]['group'] == 'HC') & 
                                        (tasks[measure]['time'] == 't1')].reset_index(drop=True)
    df_hc_t2_rt = tasks[measure][(tasks[measure]['group'] == 'HC') & 
                                        (tasks[measure]['time'] == 't2')].reset_index(drop=True)
    
    df_an = pd.DataFrame()
    df_hc = pd.DataFrame()
    df_an['correct'] = (df_an_t1_correct['iscorrect'] + df_an_t2_correct['iscorrect']) /2
    df_hc['correct'] = (df_hc_t1_correct['iscorrect'] + df_hc_t2_correct['iscorrect']) /2
    df_an['rt'] = (df_an_t1_rt['rt'] + df_an_t2_rt['rt']) /2
    df_hc['rt'] = (df_hc_t1_rt['rt'] + df_hc_t2_rt['rt']) /2
    correct_mwu = pin.mwu(df_an['correct'], df_hc['correct'])
    rt_mwu = pin.mwu(df_an['rt'], df_hc['rt'])
    mwu_models[f'{measure}_correct'] = correct_mwu
    mwu_models[f'{measure}_rt'] = rt_mwu
    
pvals = [mwu_models[model]['p-val'].values[0] for model in mwu_models.keys()]
fwe_pvals = multitest(pvals)

for key, value in enumerate(mwu_models.items()):
    value[1]['fwe_pval'] = fwe_pvals[1][key]
```
